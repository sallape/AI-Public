```python
# prompt: how to remove data columns that have <= , >= >

import re

def remove_columns_with_operators(df):
  """
  Removes columns from a DataFrame that have names containing <=, >=, >.

  Args:
      df: The DataFrame to process.

  Returns:
      A new DataFrame with the filtered columns.
  """

  # Compile a regular expression to match column names with operators
  pattern = re.compile(r"(<=|>=|>).*")

  # Filter the columns based on the regular expression
  filtered_columns = [col for col in df.columns if not pattern.match(col)]

  # Return a new DataFrame with the filtered columns
  return df[filtered_columns]

# Use the function to remove columns with operators
df_filtered = remove_columns_with_operators(data_census_income)

# Print the filtered DataFrame
print(df_filtered.head())
```
